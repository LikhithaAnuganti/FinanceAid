```python
import mysql.connector
import matplotlib.pyplot as plt
```

```python
#MySQL server details
host = "127.0.0.1"
user = "root"
password = "xxx"
database = "finaid"

# Establishing a connection to the MySQL server
try:
    connection = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database
    )

    if connection.is_connected():
        db_Info = connection.get_server_info()
        print("Connected to MySQL Server version ", db_Info)
        cursor = connection.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("Your connected to database: ", record)
        
        # DB Queries
        
        print('Query 1: Students who are admitted into top 100 Universities(simple query) \n')
        sql_select_Query = "select student_id,student_name from student where university_ranking between 1 and 100;"
        cursor = connection.cursor()
        cursor.execute(sql_select_Query)
        records = cursor.fetchall()
        for row in records:
            print(row)
            
        #Queries for Visualization:
        print('\n Visualization Query- 1: Distribution of students requirements: \n')
        sql_select_Query = "SELECT DISTINCT(requirement) FROM student"
        cursor = connection.cursor()
        cursor.execute(sql_select_Query)
        records = cursor.fetchall()
        requirements = [row[0] for row in records]

        plt.bar(requirements, [1] * len(requirements))  # Assuming a value of 1 for each requirement
        plt.xlabel("Requirements")
        plt.ylabel("Count")
        plt.title("Distinct Requirements")
        plt.show()
    
        cursor = connection.cursor()
        print('\n Visualization Query- 2: Distribution of employee roles: \n')
        sql_select_Query = "SELECT employee_role, COUNT(*) FROM employees GROUP BY employee_role"
        cursor.execute(sql_select_Query)
        records = cursor.fetchall()
        roles = []
        counts = []
        for row in records:
            roles.append(row[0])
            counts.append(row[1])

        plt.pie(counts, labels=roles, autopct='%1.1f%%', startangle=140)
        plt.title("Distribution of Employee Roles")
        plt.show()
        
        
        cursor = connection.cursor()
        print('\n Visualization Query- 3: Distribution of transactions each month: \n')
        sql_select_Query = "select MONTH(transaction_date) as transaction_month,COUNT(*) as transaction_count from transactions group by transaction_month;"
        cursor.execute(sql_select_Query)
        records = cursor.fetchall()
        months = []
        counts = []
        for row in records:
            months.append(row[0])
            counts.append(row[1])

        plt.bar(months, counts, tick_label=[f'{m}' for m in months])
        plt.xlabel("Months")
        plt.ylabel("Transaction Count")
        plt.title("Distribution of Transactions by Month")
        plt.show()
        
except mysql.connector.Error as e:
    print(f"Error: {e}")

finally:
    # Closing the connection
    if 'connection' in locals():
        connection.close()
        print("MySQL connection closed")
```
